# `bucket_report_metrics` — Métricas de regresión por bucket
Evalúa qué tan bien predice un modelo en distintos rangos de valor.

```python
import numpy as np
import pandas as pd
from apoyo.metrics import bucket_report_metrics

np.random.seed(42)
```

## Datos sintéticos — precios de inmuebles

```python
n = 2_000
real  = np.random.lognormal(mean=12, sigma=0.8, size=n)          # precios reales
noise = np.random.normal(loc=0, scale=0.18, size=n)              # error del modelo
pred  = real * np.exp(noise)                                      # predicciones

df = pd.DataFrame({
    "precio_real": real,
    "precio_pred": pred,
    "zona": np.random.choice(["Lima Norte", "Lima Sur", "Lima Este"], size=n),
})
df.head()
```

## Reporte global

```python
reporte = bucket_report_metrics(
    df, y_true="precio_real", y_pred="precio_pred",
    thr=0.25, add_metrics=True
)

reporte.style.format({
    "p_under": "{:.1%}",
    "p_preci": "{:.1%}",
    "p_over":  "{:.1%}",
    "mape":    "{:.1f}%",
    "rmse":    "{:,.0f}",
})
```

## Reporte desagregado por zona

```python
reporte_zona = bucket_report_metrics(
    df, y_true="precio_real", y_pred="precio_pred",
    thr=0.25, by=["zona"], add_metrics=True
)

reporte_zona.style.format({
    "p_under": "{:.1%}",
    "p_preci": "{:.1%}",
    "p_over":  "{:.1%}",
    "mape":    "{:.1f}%",
    "rmse":    "{:,.0f}",
})
```

## Buckets sobre el precio predicho (`bucket_on='pred'`)

```python
bucket_report_metrics(
    df, y_true="precio_real", y_pred="precio_pred",
    thr=0.25, bucket_on="pred"
)
```
