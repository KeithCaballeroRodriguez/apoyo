# SPSS / SAV — Procesamiento de encuestas
Funciones para leer archivos `.sav`, estandarizar metadatos y guardar como Parquet.

```python
import pyreadstat
import pandas as pd
from apoyo.spss import (
    estandarizacion_metadatos,
    sav_to_parquet,
    print_variable_info,
    get_variable_labels,
)
```

## Leer un archivo SAV

```python
# Reemplaza con la ruta a tu archivo
sav_file = "mi_encuesta.sav"

df, meta = pyreadstat.read_sav(sav_file)
print(f"Filas: {len(df):,}  |  Columnas: {len(df.columns)}")
df.head(3)
```

## `estandarizacion_metadatos` — Renombrar y extraer metadatos

```python
df_clean, df_variables, df_etiquetas = estandarizacion_metadatos(df, meta)

print("=== Columnas originales → estandarizadas ===")
for orig, new in zip(meta.column_names, df_clean.columns):
    print(f"  {orig!r:30} → {new!r}")
```

```python
print("\n=== Tabla de variables ===")
df_variables.head(10)
```

```python
print("\n=== Tabla de etiquetas ===")
df_etiquetas.head(10)
```

## `print_variable_info` — Ver variables con detalle

```python
# Muestra info de todas las variables
print_variable_info(df_variables, df_etiquetas)

# Filtrar por keyword
# print_variable_info(df_variables, df_etiquetas, like="edad")
```

```python
# Retorna DataFrame
get_variable_labels(df_variables)
```

## `sav_to_parquet` — Todo en una sola llamada

```python
import os
os.makedirs("output/data", exist_ok=True)
os.makedirs("output/meta", exist_ok=True)

sav_to_parquet(
    sav_file      = "mi_encuesta.sav",
    nombre_base   = "encuesta_2024",
    path_dir_data = "output/data",
    path_dir_meta = "output/meta",
)

# Genera:
#   output/data/encuesta_2024.parquet
#   output/meta/encuesta_2024.meta_columns.parquet
#   output/meta/encuesta_2024.meta_labels.parquet
print(os.listdir("output/data"))
print(os.listdir("output/meta"))
```
