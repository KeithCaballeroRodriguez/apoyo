# GCP + Base de datos
Funciones para leer/escribir Parquet en Google Cloud Storage y conectar a PostgreSQL.

> **Nota:** estas celdas requieren credenciales reales (GCP auth y `.env`). Se muestran como referencia de uso.

## GCP — Google Cloud Storage

```python
from apoyo.gcp import read_parquet_gcp, write_parquet_gcp, list_files_gcp
```

### `list_files_gcp` — Listar archivos en un bucket

```python
# Listar todos los archivos
archivos = list_files_gcp("mi-bucket/data/2024/")
print(f"{len(archivos)} archivos encontrados")
for f in archivos[:5]:
    print(" ", f)
```

```python
# Filtrar por keyword
parquets = list_files_gcp("mi-bucket/data/2024/", kwd="encuesta")
print(parquets)
```

### `read_parquet_gcp` — Leer Parquet desde GCS

```python
df = read_parquet_gcp("mi-bucket/data/2024/encuesta_q1.parquet")
print(df.shape)
df.head()
```

### `write_parquet_gcp` — Escribir Parquet a GCS

```python
import pandas as pd

df_resultado = pd.DataFrame({"id": [1, 2, 3], "valor": [100, 200, 300]})
write_parquet_gcp(df_resultado, "mi-bucket/outputs/resultado.parquet")
print("Guardado correctamente")
```

## Base de datos — PostgreSQL

```python
from dotenv import dotenv_values
from apoyo.db import create_pg_engine
import pandas as pd
```

### `create_pg_engine` — Conectar a PostgreSQL

```python
config = dotenv_values(".env")  # USER, PWD, HOST, PORT, DBNAME
engine = create_pg_engine(config)

# Leer una tabla
df = pd.read_sql("SELECT * FROM mi_esquema.mi_tabla LIMIT 5", engine)
df
```
