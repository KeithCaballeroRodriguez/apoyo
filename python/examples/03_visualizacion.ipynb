# `theme_apoyo` + `plot_dual_axis` — Visualización
Tema matplotlib consistente y gráfico de doble eje Y.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from apoyo.plots import theme_apoyo, plot_dual_axis

%matplotlib inline
theme_apoyo()   # aplica estilo globalmente
```

## `theme_apoyo` — Comparación antes / después

```python
import matplotlib as mpl

x = np.linspace(0, 10, 100)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 4))

# Sin tema
with mpl.rc_context(mpl.rcParamsDefault):
    ax1.plot(x, np.sin(x))
    ax1.set_title("Sin theme_apoyo (default matplotlib)")
    ax1.grid(True)

# Con tema
ax2.plot(x, np.sin(x), color="#1f497d")
ax2.set_title("Con theme_apoyo")

plt.tight_layout()
plt.show()
```

## `plot_dual_axis` — Ventas vs Tasa de conversión

```python
np.random.seed(0)
meses = pd.date_range("2023-01", periods=24, freq="MS")
df = pd.DataFrame({
    "fecha":      meses,
    "ventas":     np.cumsum(np.random.randint(50_000, 200_000, 24)) + 500_000,
    "conversion": np.random.uniform(0.05, 0.25, 24),
})

fig, (ax1, ax2) = plot_dual_axis(
    df,
    x      = "fecha",
    y1     = "ventas",
    y2     = "conversion",
    label1 = "Ventas (S/.)",
    label2 = "Tasa de conversión",
)
ax2.yaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: f"{y:.0%}"))
plt.title("Ventas vs Tasa de conversión — 2023-2024")
plt.show()
```

## `plot_dual_axis` con `include_zero=True`

```python
fig, _ = plot_dual_axis(
    df,
    x            = "fecha",
    y1           = "ventas",
    y2           = "conversion",
    label1       = "Ventas (S/.)",
    label2       = "Tasa de conversión",
    include_zero = True,
    color1       = "#c0392b",
    color2       = "#27ae60",
)
plt.title("Con include_zero=True")
plt.show()
```
