# `clean_string` — Normalización de texto
Elimina acentos, caracteres especiales y convierte a `snake_case`.

```python
from apoyo.text import clean_string
```

## Casos básicos

```python
casos = [
    "  Hola Mundo!  ",
    "Ñoño & Compañía S.A.",
    "P1. ¿Cuál es su edad?",
    "REGIÓN_NORTE - Lima Metropolitana",
    "precio_m2 (S/.)",
    "NivelSocioEconómico",
]

for txt in casos:
    print(f"{txt!r:45} → {clean_string(txt)!r}")
```

## Uso en renombrar columnas de un DataFrame

```python
import pandas as pd

df = pd.DataFrame(columns=[
    "Región", "Código NSE", "P1. ¿Edad?", "Precio (S/.)", "¿Tiene deuda? (Sí/No)"
])

df.columns = [clean_string(c) for c in df.columns]
print(df.columns.tolist())
```
